### FIRST MAKE A COPY OF THE ORIGINAL DATASETS

```python
movie_copy = nmo.copy()
movie_copy.head()
```

```python
tv_copy=ntv.copy()
tv_copy.head()
```

AS the output suggests the duration column in tvshow has no unique value it only has 1 season and the duration column in movies is blank. So it is safe to say that it is useless for further consideration.** SO WE DELETE THIS FROM BOTH THE DATASETS**

```python
movie_copy = movie_copy.drop('duration', axis=1)
movie_copy.head()
```

```python
tv_copy = tv_copy.drop('duration', axis=1)
tv_copy.head()
```

DROPPING THE DESCRIPTION COLUMNS FROM BOTH

```python
movie_copy = movie_copy.drop('description', axis=1)
movie_copy.head()
```

```python
tv_copy = tv_copy.drop('description', axis=1)
tv_copy.head()
```

```python
tv_copy.fillna({'country':'missing','director':'missing','cast':'missing','genres':'missing'},inplace=True)
```

```python
tv_copy.head(2)
```

```python
tv_copy.isnull().sum()
```

```python
movie_copy.fillna({'country':'missing','director':'missing','cast':'missing','genres':'missing'},inplace=True)
```
